#### Chart - 1 : Distribution of Products Across Categories

````python
# Chart - 1 visualization code
plt.figure(figsize=(12,8))
df_2['Category'].value_counts().plot(kind='bar')

plt.title('Number of Products in Each Category')
plt.xlabel('Category')
plt.ylabel('Number of Products')
plt.show()
````

##### 1. Why did you pick the specific chart?

*A bar chart is the best choice for comparing the number of products across different categories. It clearly highlights categories with the highest and lowest product counts.*




##### 2. What is/are the insight(s) found from the chart?

*The inventory is not evenly distributed across categories.*

*Categories like Cooking Essentials and Munchies have the highest inventory while meat,fish and eggs have lowest inventory.*

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

#### Chart - 11 : Out-of-Stock Products by Category

````python
# Chart - 11 visualization code

out_stock = df_2[df_2['outOfStock'] == True]

category_out = out_stock['Category'].value_counts()

plt.figure(figsize=(10,10))

plt.pie(category_out,
        labels=category_out.index,
        autopct='%1.1f%%',
        startangle=90)

plt.title('Out-of-Stock Products by Category')

plt.show()
````

Pie chart provides a good visual representation in the case of multiple categories and hence can be used for showing the respective perecentages of each category.

#### Chart - 12 : Products Above and Below Average Selling Price

````python
# Chart - 12 visualization code
avg_price = df_2['discountedSellingPrice'].mean()

price_group = df_2['discountedSellingPrice'].apply(
    lambda x: 'Above Average Price' if x > avg_price else 'Below Average Price'
)
price_group.value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    figsize=(8,8),
    startangle=90
)
plt.title('Products Above and Below Average Selling Price')
plt.ylabel('')
plt.show()
````

##### 1. Why did you pick the specific chart?

Pie chart gives a visual representation of the percentage of products above and below average price.

#### Chart - 13 : Average Selling Price by Category and Stock Status

#### Chart - 14 - Correlation Heatmap

````python
# Correlation Heatmap visualization code
plt.figure(figsize=(8,6))
sns.heatmap(
    df_2.corr(numeric_only=True),
    annot=True,
    cmap='coolwarm',
    fmt='.2f',
    linewidths=0.5
)

plt.title('Correlation Heatmap of Numerical Variables')

plt.show()
````

##### 1. Why did you pick the specific chart?

A correlation heatmap is used to visualize the relationship between numerical variables. It quickly identifies strong positive, negative, or weak correlations, helping understand how different variables influence one another.

##### 2. What is/are the insight(s) found from the chart?

MRP and Discounted Selling Price show a strong positive correlation, indicating that products with higher MRPs generally have higher selling prices after discounts.

Discount Percentage has a negative correlation with Discounted Selling Price, meaning larger discounts usually reduce the final selling price.

Available Quantity and Weight have weak correlations with other variables, suggesting they do not significantly influence product pricing.
